```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('data_full.csv')
```

```python
splits=['[','[[',']]',']',"'"]
criteria=list(data['criteria'])
criteria_list=list()
#print(str(criteria))
#for i in splits:
    #criteria=str(criteria).split(i)
for j in range(len(criteria)):
    try: 
        for i in splits:
            criteria[j]=criteria[j].replace(i, '')
        criteria[j]=criteria[j].split(',')
        criteria_list+=criteria[j]
    except: 
        None
```

```python
values=pd.DataFrame(criteria_list)[0].value_counts()[0:500]
```
